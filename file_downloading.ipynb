```python
import os
import gdown

folder_url = "https://drive.google.com/drive/folders/1p5s4APqO8szykm3LLf0jDp8q2D89H4o0?usp=drive_link"
output_dir = "data/raw"

os.makedirs(output_dir, exist_ok=True)

# Запуск завантаження
gdown.download_folder(url=folder_url, output=output_dir, quiet=False, use_cookies=False)
```

```python
import os
import re

# Шлях до папки з файлами
folder_path = "data"

# Проходимо по всіх файлах у папці
for filename in os.listdir(folder_path):
    # Шукаємо рік у назві, наприклад "Odata2021File.csv"
    match = re.search(r'Odata(\d{4})', filename)
    if match:
        year = match.group(1)
        old_path = os.path.join(folder_path, filename)
        new_filename = f"{year}.csv"
        new_path = os.path.join(folder_path, new_filename)

        # Перейменування
        os.rename(old_path, new_path)
        print(f"✅ {filename} → {new_filename}")
```
