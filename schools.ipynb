```python
import pandas as pd
import numpy as np
```

```python
def select_col(df):
    ball100 = []
    for col in df.columns:
        if 'score_100' in col:
            ball100.append(col)
    df = df[['education_org_type'] + ball100]
    return df
```

```python
grades_only = ['ukrainian_score_100', 'history_score_100', 'math_score_100',
               'physics_score_100', 'chemistry_score_100', 'biology_score_100',
               'geography_score_100', 'english_score_100', 'french_score_100',
               'german_score_100', 'spanish_score_100']
```

# EIT

```python
df_2019 = pd.read_csv('transformed_data/2019.csv', encoding='', low_memory=False, sep=',')
df_2019 = select_col(df_2019)

df_2020 = pd.read_csv('transformed_data/2020.csv', encoding='', low_memory=False, sep=',')
df_2020 = select_col(df_2020)

df_2021 = pd.read_csv('transformed_data/2021.csv', encoding='', low_memory=False, sep=',')
df_2021 = select_col(df_2021)

EIT = pd.concat([df_2019, df_2020, df_2021], ignore_index=True)
```

```python
EIT[grades_only] = EIT[grades_only].replace(0, np.nan)
EIT['Median Score'] = EIT[grades_only].median(axis=1, skipna=True)
```

```python
eit_by_education = EIT.groupby('education_org_type').agg(
    median_score=('Median Score', 'median'),
    count_of_students=('education_org_type', 'count')
).reset_index()
eit_by_education.rename(columns={
    'median_score': 'Median Score'
}, inplace=True)

eit_by_education = eit_by_education.sort_values('Median Score', ascending=False)
eit_by_education
```
